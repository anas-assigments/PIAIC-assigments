```python
import numpy as np 
import matplotlib.pylab as plt

def plti(im, h=8, **kwargs):
    """
    Helper function to plot an image.
    """
    y= im.shape[0]
    x= im.shape[1]
    w= (y/x)*h
    plt.figure(figsize=(w,h))
    plt.imshow(im, interpolation="none",**kwargs)
    plt.axis("off")
    
%matplotlib inline
im = plt.imread("air.jpg")
print(im.shape)
#plti(im)
im = im[:1080:,:1920:]
plti(im)
print(im.shape)
```

```python
import numpy as np
from PIL import Image
def resize(*images):
    resized_img = []
    for i in images:
        img = Image.open(i)
        new_img = img.resize((300,300))
        resized_img_array = np.array(new_img)
        resized_img.append(resized_img_array)
        
    return resized_img
        
list1 = resize("/home/joker/assigment_imgs/05.jpeg")
print(list1)
```
